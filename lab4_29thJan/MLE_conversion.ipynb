```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(33)

true_p = 0.7
sample_size = [10,20,50,100,500,1000]
p_mle_values = []

for n in sample_size:
    data = np.random.binomial(1, true_p, n)
    p_mle = data.mean()
    p_mle_values.append(p_mle)


plt.plot (sample_size, p_mle_values, marker = "o")
plt.axhline(true_p, linestyle = "--")
plt.xlabel("Sample size")
plt.ylabel("p_mle_values")
plt.title("MLE concergence with")
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(33)

true_mu = 2
true_sigma = 1

sample_size = [10, 20, 50, 100, 500, 1000]

mu_mle_values = []
sigma_mle_values = []

for n in sample_size:
    data = np.random.normal(true_mu, true_sigma, n)
    
    mu_mle = data.mean()
    sigma_mle = data.std(ddof=0)  
    
    mu_mle_values.append(mu_mle)
    sigma_mle_values.append(sigma_mle)


plt.plot (sample_size, mu_mle_values, marker = "o")
plt.axhline(true_mu, linestyle = "--")
plt.xlabel("Sample size")
plt.ylabel("mu_mle_values")
plt.title("MLE concergence with")
plt.show

plt.plot (sample_size, sigma_mle_values, marker = "^")
plt.axhline(true_sigma, linestyle = "--")
plt.xlabel("Sample size")
plt.ylabel("sigma_mle_values")
plt.title("MLE concergence with")
plt.show
```
