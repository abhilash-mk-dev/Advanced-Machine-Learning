```python
import numpy as np

np.random.seed(33)

true_p = 0.7
data = np.random.binomial(1, true_p, 100)

print(f"Data with {true_p}:",data)
print(data.sum())
```

Here we are assuming that the data follows a bernoulli
P(X = 1) = p

```python
def log_like(p,data):
    return np.sum( data* np.log(p) + (1 - data) * np.log(1-p))

```

```python
import matplotlib.pyplot as plt

p_values = np.linspace(0.01, 0.99, 100)
ll_values = [log_like(p , data) for p in p_values]
```

```python
plt.plot(p_values, ll_values)
```
