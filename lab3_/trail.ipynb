```python
from sklearn.datasets import load_breast_cancer
```

```python
# data explorartion
# train a linear model 
# train a plonomial 
# check accuracy

```

```python
data = load_breast_cancer()
X = data.data
y = data.target
type(X)
```

```python
X 
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from visual import plot_2d_data
from sklearn.svm import SVC

linear_svm = SVC(kernel='linear')
linear_svm.fit(X, y)
linear_y_pred = linear_svm.predict(X)
linear_accuracy = accuracy_score(y, linear_y_pred)
print(f"Accuracy linear SVM: {linear_accuracy}")
plot_2d_data(X, linear_y_pred, title="Linear SVM prediction")


```

```python
ploy_svm = SVC(kernel="poly", degree=2)
ploy_svm.fit(X,y)
ploy_y_pred = ploy_svm.predict(X)
ploy_accuracy = accuracy_score(y, ploy_y_pred)
print("Poly accuracy:", ploy_accuracy)
plot_2d_data(X,ploy_y_pred, title = "Ploynomial SVM")
```
